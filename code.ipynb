```python
from PIL import Image, ImageDraw

# 定义图片宽高
width = 4000
height = 2000

# 定义边框的宽度
border_width = 4

# 创建图像对象
image = Image.new("RGB", (width + 2 * border_width, height + 2 * border_width), "black")
draw = ImageDraw.Draw(image)

# 绘制边框
draw.rectangle([(0, 0), (width + 2 * border_width - 1, height + 2 * border_width - 1)], outline="blue", width=border_width)

# 保存图像
image.save("bordered_image.png")
```
